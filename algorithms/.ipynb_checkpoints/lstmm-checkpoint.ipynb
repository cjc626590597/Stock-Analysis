```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

```python
# Initialising the RNN
def build_model(indicators, units=32, activation='tanh', learning_rate=0.001, dropout_rate=0.1):
    model = Sequential()

    # Adding the first LSTM layer and some Dropout regularisation
    model.add(LSTM(units = units, activation=activation, return_sequences = True, input_shape = (60, len(indicators))))
    model.add(Dropout(dropout_rate))
    # Adding a second LSTM layer and some Dropout regularisation
    model.add(LSTM(units = units, activation=activation, return_sequences = True))
    model.add(Dropout(dropout_rate))
    # Adding a third LSTM layer and some Dropout regularisation
    model.add(LSTM(units = units, activation=activation))
    model.add(Dropout(dropout_rate))
    
    # Adding the output layer
    model.add(Dense(units = 1, activation=activation))

    # Compiling the RNN
    model.compile(optimizer = tf.optimizers.Adam(learning_rate=learning_rate), loss = 'mean_squared_error')
    return model
```

```python
def transform_60(indicators, X_data, y_data):
    # Creating a data structure with 60 timesteps and 1 output
    x = []
    y = []
    for i in range(59, X_data.shape[0]):
        x.append(np.array(X_data.iloc[i-59:i+1][indicators]))
        y.append(np.array(y_data.iloc[i][horizons][0]))
    x, y = np.array(x), np.array(y)
    return x, y
```

```python
def lstm_train(indicators, X_train, y_train, X_test, y_test, model=None):
    x = X_train.iloc[-59:]
    y = y_train.iloc[-59:]
    X_test = pd.concat([x, X_test], axis=0)
    y_test = pd.concat([y, y_test], axis=0)
    
    X_train, y_train = transform_60(indicators, X_train, y_train)
    X_test, y_test = transform_60(indicators, X_test, y_test)
    
    if model == None:
        model = build_model(indicators)
    model.fit(X_train, y_train, epochs = 20, batch_size = 32)
    
    preds_train = model.predict(X_train)
    preds_test = model.predict(X_test)
    return preds_train, preds_test
```

```python
# Fitting the RNN to the Training set
model = build_model(technical_indicators, units=64, activation='tanh', dropout_rate=0.2)
preds_train, preds_valid = lstm_train(technical_indicators, X_train_split, y_train_split, X_valid_split, y_valid_split, model=model)
print_score("Technical Indicators", horizons, "NN", y_valid_split, preds_valid)
plt.figure(figsize=(12, 8))
plot_prediction("technical_indicators", horizons, "LSTM", y_scaler, y_train_split.iloc[59:], y_valid_split, preds_train, preds_valid)
```

```python
rms_mean = []
neuron_range = [8, 16, 32, 64, 128]

# C = 1/lamda: larger C causes overfitting, smaller C causes underfitting
for neuron in neuron_range:
    
    rms = []
    
    model = build_model(technical_indicators, units=neuron)
    _, preds_valid = lstm_train(technical_indicators, X_train_split, y_train_split, X_valid_split, y_valid_split, model)
    rms.append(score(horizons, y_valid_split, preds_valid))
    model = build_model(fundamental_indicators, units=neuron)
    _, preds_valid = lstm_train(fundamental_indicators, X_train_split, y_train_split, X_valid_split, y_valid_split, model)
    rms.append(score(horizons, y_valid_split, preds_valid))
    model = build_model(combined_indicators, units=neuron)
    _, preds_valid = lstm_train(combined_indicators, X_train_split, y_train_split, X_valid_split, y_valid_split, model)
    rms.append(score(horizons, y_valid_split, preds_valid))
        
    print("neuron:"+str(neuron)+", mean:"+str(np.mean(rms)))
        
    rms_mean.append(np.mean(rms))
```

```python
# Plot
plt.figure(figsize = (10, 6))
plt.plot(neuron_range, rms_mean, linestyle = '-.', linewidth = 2, color = 'r', alpha = 0.6, marker = 'o', markersize = 4)
plt.xlabel('Number of Hidden Neurons')
plt.ylabel('RMSE')
plt.show
```
