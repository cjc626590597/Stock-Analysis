```python
def RMSE(indicators, model, y_test, preds_test):
    # checking the results (RMSE value)
    rms=np.sqrt(np.mean(np.power((np.array(y_test['Close_after_1_day'])-preds_test),2)))
    print('\nRMSE using {} to Predict Price by {}'.format(indicators, model))
    print(rms)
```

```python
def score(y_test, preds_test):
    return np.sqrt(np.mean(np.power((np.array(y_test['Close_after_1_day'])-preds_test),2)))
```

```python
def inverse_transform(scaler, y_train, y_test, preds_train, preds_test):
    y_train.loc[: ,horizons] = scaler.inverse_transform(y_train.loc[: ,horizons]).copy()
    y_test.loc[: ,horizons] = scaler.inverse_transform(y_test.loc[: ,horizons]).copy()
    preds_train = scaler.inverse_transform(preds_train)
    preds_test = scaler.inverse_transform(preds_test)
    return y_train, y_test, preds_train, preds_test
```

```python
def plot_prediction(indicators, model, y_train, y_test, preds_train, preds_test):
    #plot
    preds_train = pd.DataFrame(preds_train, columns = ['Predictions'], index = y_train.index)
    preds_test = pd.DataFrame(preds_test, columns = ['Predictions'], index = y_test.index)
    plt.plot(y_train['Close_after_1_day'], label='Close Price', color='g')
    plt.plot(y_test[['Close_after_1_day']], label='Actual Price', color='b')
    plt.plot(preds_train['Predictions'], label='{} Predicted Price'.format(model), color='r')
    plt.plot(preds_test['Predictions'], label='{} Predicted Price'.format(model), color='r')
    plt.title('Using {} to Predict Price by {}'.format(indicators, model))
    plt.legend()
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
# create model
model = Sequential()
model.add(LSTM(units=30, return_sequences=True, input_shape=(x_train[technical_indicators].shape[1],1)))
model.add(Dropout(0.2)) 
model.add(LSTM(units = 50, return_sequences = True)) 
model.add(Dropout(0.2)) 
model.add(LSTM(units = 70, return_sequences = True)) 
model.add(Dropout(0.2))
model.add(Dense(1))

model2 = Sequential()
model2.add(LSTM(units=50, return_sequences=True, input_shape=(x_train[technical_indicators].shape[1],1)))
model2.add(LSTM(units=50))
model2.add(Dense(1)) 

model3 = Sequential()
model3.add(LSTM(units=50, return_sequences=True, input_shape=(x_train[technical_indicators].shape[1],1)))
model3.add(LSTM(units=50))
model3.add(Dense(1))
 
# Compile model
 
model.compile(loss='mean_squared_error',optimizer= 'adam')
model2.compile(loss='mean_squared_error',optimizer= 'adam')
model3.compile(loss='mean_squared_error',optimizer= 'adam')
```
