```python
def RMSE(indicators, model, y_test, preds_test):
    # checking the results (RMSE value)
    rms=np.sqrt(np.mean(np.power((np.array(y_test['Close_after_1_day'])-preds_test),2)))
    print('\nRMSE using {} to Predict Price by {}'.format(indicators, model))
    print(rms)
```

```python
def score(y_test, preds_test):
    return np.sqrt(np.mean(np.power((np.array(y_test['Close_after_1_day'])-preds_test),2)))
```

```python
def inverse_transform(scaler, y_train, y_test, preds_train, preds_test):
    train = y_train.copy()
    test = y_test.copy()
    train.loc[: ,horizons] = scaler.inverse_transform(y_train.loc[: ,horizons]).copy()
    test.loc[: ,horizons] = scaler.inverse_transform(y_test.loc[: ,horizons]).copy()
    preds_train = preds_train.reshape(y_train.shape[0], 1)
    preds_test = preds_test.reshape(y_test.shape[0], 1)
    preds_train = scaler.inverse_transform(preds_train)
    preds_test = scaler.inverse_transform(preds_test)
    return train, test, preds_train, preds_test
```

```python
def plot_prediction(indicators, model, scaler, y_train, y_test, preds_train, preds_test):
    #plot
    y_train, y_test, preds_train, preds_test = inverse_transform(scaler, y_train, y_test, preds_train, preds_test)
    preds_train = pd.DataFrame(preds_train, columns = ['Predictions'], index = y_train.index)
    preds_test = pd.DataFrame(preds_test, columns = ['Predictions'], index = y_test.index)
    plt.plot(y_train['Close_after_1_day'], label='Close Price', color='g')
    plt.plot(y_test[['Close_after_1_day']], label='Actual Price', color='b')
    plt.plot(preds_train['Predictions'], label='{} Predicted Price'.format(model), color='r')
    plt.plot(preds_test['Predictions'], color='r')
    plt.title('Using {} to Predict Price by {}'.format(indicators, model))
    plt.legend()
```

```python
# Normalizing the data set 
x_scaler = MinMaxScaler(feature_range=(0, 1))
x_scaler.fit_transform(x_all_one[combinated_indicators])
y_scaler = MinMaxScaler(feature_range=(0, 1))
y_scaler.fit_transform(y_all_one.loc[: ,horizons])
x_scaler
```

```python
x_train_one[combinated_indicators] = x_scaler.transform(x_train_one[combinated_indicators]).copy()
x_test_one[combinated_indicators] = x_scaler.transform(x_test_one[combinated_indicators]).copy()
x_train_one.head()
```

```python
y_train_one.loc[: ,horizons] = y_scaler.transform(y_train_one.loc[: ,horizons]).copy()
y_test_one.loc[: ,horizons] = y_scaler.transform(y_test_one.loc[: ,horizons]).copy()
y_test_one.tail()
```

```python
preds_train = []
preds_test = []
plt.figure(figsize=(16,8))

preds_train, preds_test = lstm_train(technical_indicators, x_train_one, y_train_one, x_test_one, y_test_one)
RMSE("Technical Indicators", "RNN", y_test_one, preds_test)
plt.subplot(3,2,1)
plot_prediction("Technical Indicators", "RNN", y_scaler, y_train_one.iloc[59:], y_test_one, preds_train, preds_test)

svr.fit(x_train_one[technical_indicators], y_train_one['Close_after_1_day'])
preds_train = svr.predict(x_train_one[technical_indicators])
preds_test = svr.predict(x_test_one[technical_indicators])
RMSE("Technical Indicators", "SVR", y_test_one, preds_test)
plt.subplot(3,2,2)
plot_prediction("Technical Indicators", "SVR", y_scaler, y_train_one, y_test_one, preds_train, preds_test)

preds_train, preds_test = lstm_train(fundamental_indicators, x_train_one, y_train_one, x_test_one, y_test_one)
RMSE("Fundamental Indicators", "NN", y_test_one, preds_test)
plt.subplot(3,2,3)
plot_prediction("Fundamental Indicators", "RNN", y_scaler, y_train_one.iloc[59:], y_test_one, preds_train, preds_test)

svr.fit(x_train_one[fundamental_indicators], y_train_one['Close_after_1_day'])
preds_train = svr.predict(x_train_one[fundamental_indicators])
preds_test = svr.predict(x_test_one[fundamental_indicators])
RMSE("Fundamental Indicators", "SVR", y_test_one, preds_test)
plt.subplot(3,2,4)
plot_prediction("Fundamental Indicators", "SVR", y_scaler, y_train_one, y_test_one, preds_train, preds_test)

preds_train, preds_test = lstm_train(combinated_indicators, x_train_one, y_train_one, x_test_one, y_test_one)
RMSE("Combinated Indicators", "NN", y_test_one, preds_test)
plt.subplot(3,2,5)
plot_prediction("Combinated Indicators", "RNN", y_scaler, y_train_one.iloc[59:], y_test_one, preds_train, preds_test)

svr.fit(x_train_one[combinated_indicators], y_train_one['Close_after_1_day'])
preds_train = svr.predict(x_train_one[combinated_indicators])
preds_test = svr.predict(x_test_one[combinated_indicators])
RMSE("Combinated Indicators", "SVR", y_test_one, preds_test)
plt.subplot(3,2,6)
plot_prediction("Combinated Indicators", "SVR", y_scaler, y_train_one, y_test_one, preds_train, preds_test)

plt.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=.5,hspace=.5)
```
