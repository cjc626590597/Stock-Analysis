```python
import pandas as pd
# Get the stock symbol of S&P 100 and save the results
symbols = pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')[2].Symbol
sectors = pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')[2].Sector
df = pd.DataFrame({'Symbol': symbols, 'Sector': sectors})
df.to_csv('../data/sp100tickers.csv', sep=',', index=False)
df
```
